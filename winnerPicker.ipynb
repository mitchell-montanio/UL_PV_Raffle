```python
## mandatory libraries for this code to work: xlrd, openpyxl, pandas, numpy, sklearn
## these can all be installed using pip on the command line
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
```

```python
## read the excel file into a dataframe. Name of file is assumed to be EntrantsList.xlsx, but this can be modified if necessary
df = pd.read_excel('EntrantsList.xlsx')
```

```python
## create a new dataframe that shuffles the old dataframe. Reset the index so the shuffling is effective
df_shuf = shuffle(df).reset_index(drop=True)
```

```python
## pick a random winner
winning_index = np.random.randint(len(df))
```

```python
## assign the winner to this variable
winner = df_shuf.loc[winning_index]
```

```python
## print the winner's details
print(winner)
```
